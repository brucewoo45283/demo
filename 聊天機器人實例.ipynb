## (一) 使用者在建立物件時，可以給予機器人的年齡(age)與性別(gender)¶

```python
class Robert(object):               #class 類別名稱(物件object)，類別為Robert
    def __init__(self, age, gender):# 定義封裝的變數: def 建構子(內部屬性, 參數)
        self.age=age                #定義封裝的函式: 內部屬性.參數=參數
        self.gender=gender
        
```

### 創立物件i

```python
i=Robert("27", "M")
```

```python
print(i)
```

## (二) 為機器人建立函式 getRobotProperty: 可以列印出機器人的年齡與性別
### 1.與類和例項無繫結關係的function都屬於函式（function）；
### 函式（FunctionType）
### 函式是封裝了一些獨立的功能，可以直接呼叫，能將一些資料（引數）傳遞進去進行處理，然後返回一些資料（返回值），也可以沒有返回值。可以直接在模組中進行定義使用。 所有傳遞給函式的資料都是顯式傳遞的。
### 2.與類和例項有繫結關係的function都屬於方法（method）。
### 方法和函式類似，同樣封裝了獨立的功能，但是方法是只能依靠類或者物件來呼叫的，表示針對性的操作。
### 方法中的資料self和cls是隱式傳遞的，即方法的呼叫者；
### 方法可以操作類內部的資料
### 簡單的說，函式在python中獨立存在，可直接使用的，而方法是必須被別人呼叫才能實現的。
### 靜態方法除外（與類和物件均無關，通過類名和物件名均可被呼叫，屬函式）
### 在模組中實現的函式只要匯入這個模組的這個函式既可以任意使用了，但是在類中宣告的必須匯入這個類之後然後再通過建立例項或者類名來呼叫。可以說直接在模組中宣告的是更普遍廣泛性的函式，而在類中宣告的方法一般是專屬於一類事物特有的

# 物件 = 屬性attribute + 方法method

```python
class Robert(object):                #類別(物件):用class來建立一個模板，用模板來產生大量物件                     
    def __init__(self, age, gender): #定義屬性(attribute)::封裝的變數有age, gender
        self.age=age                
        self.gender=gender
    def getRobotProperty(self):     #在物件中，我建立方法(method)為getRobotProperty(self),self為指向自己
        print("我的年齡是 {}, 我的性別是: {}".format(self.age, self.gender)) 
```

```python
i=Robert("27", "男")
```

```python
print(i)
```

### ▸ 每個物件都是某個類別 (Class) 所產生的一個實例 (Instance)，就好像工廠裡一個模具可以產生許多相同規格的產品一般：模具就是類別，產品就是實例
### ▸ 類別裡可包含屬性 (Attribute) 及方法 (Method)
### ✶ 屬性：儲存物件的資訊，也就是變數
### ✶ 方法：操作物件的資訊，也就是函式

###  呼叫、執行方法 
### <物件>.<方法>() 

```python
i.getRobotProperty()
```

### [Python教學]Python Unpacking實用技巧分享
https://www.learncodewithmike.com/2019/12/python-unpacking.html

```python
answer_dic={}
with open("cdc_answer.tsv", "r", encoding="utf-8") as f:
    for line in f.readlines():
        #print(line)  #先看迴圈內長怎樣，再決定怎麼寫code
        keyword, answer = line.strip().split("\t")     #用tab分隔
        #print(keyword)
        answer_dic[keyword]=answer
        
```

# 以for迴圈，建立問答集

```python
q=input("請輸入疾病的關鍵字:")
```

```python
for keyword in answer_dic:
    if keyword in q:
        print(answer_dic[keyword])
        break
```

# 建立函式 getMessage
## 把字典直接建立在建構子

```python
q=input("請輸入疾病的關鍵字:")
```

```python
class Robert(object):                                   
    def __init__(self, age, gender): 
        self.age=age                
        self.gender=gender
        self.answer_dic={}
        with open("cdc_answer.tsv", "r", encoding="utf-8") as f:
            for line in f.readlines():
        #print(line)  #先看迴圈內長怎樣，再決定怎麼寫code
                keyword, answer = line.strip().split("\t")     #用tab分隔
        #print(keyword)
                self.answer_dic[keyword]=answer
    def getRobotProperty(self):     
        print("我的年齡是 {}, 我的性別是: {}".format(self.age, self.gender)) 
    def getMessage(self, question):
        for keyword in self.answer_dic:
            if keyword in question:
                return self.answer_dic[keyword]
        return "不懂你的意思，請重新輸入"
        
```

```python
i=Robert("27", "男")
print(i)
```
