```python
### 根據多個欄位做統計
df.groupby(['author', 'source'])['author'].count().head() #發覺都是kickingbird作者發的新聞
```

```python
### 根據月份、source做分組統計
###先取出月份，並新增一個month欄位
df["month"]=df["dt"].dt.month
df.head()
```

```python
cnt=df.groupby(['month', 'source'])['title'].count().to_frame() #.to_frame()轉成DayaFrame 才能做樞紐
cnt
```
