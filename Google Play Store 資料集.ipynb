# 排除

```python
df1=df[df["Rating"]<=5]
```

```python
df1
```

```python
df1["Rating"].nlargest(100).mean()
```

```python
df1["Installs"]  #字串
```

```python
#字串轉數字
```

```python
s="10,000+"
```

```python
int(s.replace(",","").replace("+",""))
```

```python
def k(i):
    res=int(i.replace(",","").replace("+",""))
    return(res)
```

```python
df1["Installs"]=df1["Installs"].map(k)
```

```python
df1
```

```python
df1["Installs"].mean()
```

# 安裝數量>10000有幾個

```python
df1[df1["Installs"]>100000]
```
