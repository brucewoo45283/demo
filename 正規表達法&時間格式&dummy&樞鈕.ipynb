```python
#sol(1):寫出自訂函式
def normalizePrice(ele):
    res= int(ele.replace("元/月", "").replace(",",""))
    return res
normalizePrice("27,000元/月")
```

```python
#把自訂函式套用到Series (df["price"]) 上的每個元素 : map()
df["price"]=df["price"].map(normalizePrice)
```

# 時間回溯:timedelta

```python
from datetime import timedelta
#回溯一天
ta - timedelta(days = 1)
```

```python
# 回溯前十天的資料
for i in range(11):
    tt = ta - timedelta(days = i)  
    print(tt.strftime('%Y/%m/%d')) #轉成字串
```

```python
#example:每個月的最後一周的禮拜五要過帳，我要列出該年所有日子
st = datetime.strptime('2021-01-01', '%Y-%m-%d')  #列出本年第一天，起始日
st
ary = []
```

```python
# 每個月最後一周的星期五
for i in range(365):
    dt = st + timedelta(days = i)
    date  = dt.strftime('%Y-%m-%d')
    weekday  = int(dt.strftime('%w'))
    week  = int(dt.strftime('%W'))
    month = int(dt.strftime('%m'))
    ary.append({'date':date, 'weekday':weekday, 'week':week, 'month':month})
```

```python
# 放進DataFrame
df2 = pandas.DataFrame(ary)
df3 = df2[df2['weekday'] == 5] #篩選
df3
```

```python
#取該月的最後一個禮拜五
df3.groupby('month')['week'].transform('max')
```

```python
df3[df3.groupby('month')['week'].transform('max') == df3['week']]
```
