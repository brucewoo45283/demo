# 正確做法
### request header : 一般的使用者，在連線的時候，會發送的資訊
### 底下的user agent:使用者的作業系統、瀏覽器，此步驟會讓ptt認為我們是正常使用者
### 建立一個Request物件，附加Request Headers資訊(模仿人類)

```python
url="https://www.ptt.cc/bbs/AllTogether/index.html"
request=req.Request(url, headers={
  "User-Agent" :" Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36" })
with req.urlopen(request) as response:  #利用request物件來打開網址，不要直接丟網址
    data=response.read().decode("utf-8")
print(data)
```

### 改成回傳"上一頁"連結

```python
def ptt(url):
    request1=req.Request(url, headers={
    "cookie":"over18=1",
    "User-Agent" :" Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36" })
    with req.urlopen(request1) as response:  #利用request物件來打開網址，不要直接丟網址
        data1=response.read().decode("utf-8")
    import bs4
    root1=bs4.BeautifulSoup(data1, "html.parser")
    titles1=root1.find_all("div", class_="title")
    for title1 in titles1:
        if title1.a != None: #!
            print(title1.a.string)
    nextlink=root1.find("a", string="‹ 上頁")
    return nextlink["href"]
```

### 呼叫函式

```python
pageURL="https://www.ptt.cc/bbs/basketballTW/index.html"
pageURL=ptt(pageURL) #呼叫ptt()的函式，把pageURL丟進函式運算，得出一個pageURL的物件，return 上一頁的連結
print(pageURL)
```

### 得到完整"上一頁"的網址：函式只回傳"/bbs/basketballTW/index3087.html"，要自己接上"https://www.ptt.cc"

```python
pageURL="https://www.ptt.cc/bbs/AllTogether/index.html" #代表第一頁
pageURL="https://www.ptt.cc"+ ptt(pageURL) 
print(pageURL)
```
