```python
import pandas as pd
#建立DataFrame-1，後面添加colums name
df=pd.DataFrame([['frank','M',29],['mary','F',23],['tom','M',35],['ted','M',33],['jean','F',21],['lisa','F',20]], columns=['name','gender','age'])
df
```

```python
#建立DataFrame-2 字典的形式(如果有NA值)，以後如果有字典的資料，請丟到DataFrame操做資料
df1=pd.DataFrame([{"name":"frank", "gender":"M", "age":29},\
                 {"name":"mary", "gender":"F", "age":23},\
                 {'name':'tom','gender':"NA",'age':35},\
                 {'name':'ted','gender':'M','age':33},\
                 {'name':"NA",'gender':'F','age':21},\
                 {'name':'lisa','gender':'F','age':20}])

df1
```

```python
#取值
#取出欄位名的所有值，包在list裡
df[["age", "name"]]
```

```python
#取出某欄位某特定值，的所有列
df[df["gender"]=="M"]
```

```python
#欄位取值:交集與連集
#交集:性別為男且年齡30歲以上
df[(df["gender"]=="M") & (df["age"]>=30)]
```

```python
#聯集:性別為男或年齡30歲以上
df[(df["gender"]=="M") | (df["age"]>=30)]
```

```python
#刪除欄位
df=df.drop("employee", 1)  #1:刪掉整欄 0:刪掉整列
df
```

```python
#刪除
#刪除
df["employee"]=True
df=df.drop("employee", 1)  #1:刪掉整欄 0:刪掉整列
df
```

```python
#刪掉重複值
df2 = pd.DataFrame([['a', '1'], ['a', '2'], ['b', '3'], ['b', '3'],['c', '2'], ['d', '20']])
df2
```

```python
#把重複的資料拿掉 [b,3]
df3=df2.drop_duplicates()
df3
```

```python
#根據某欄位，刪重複的資料
df4=df2.drop_duplicates(0,keep='last')  #根據欄位0，保留最後一筆重複值
df4
```
